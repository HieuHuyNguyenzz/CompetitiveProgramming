# Binary Search
Tìm kiếm nhị phân là một thuật toán được sử dụng để tìm kiếm phần tử trong một danh sách đã được sắp xếp. Thuật toán này hoạt động bằng cách lặp đi lặp lại quá trình cắt giảm nửa phần không cần thiết của danh sách cho đến khi tìm ra phần tử cần tìm hoặc xác định rằng phần tử đó không có trong danh sách.

Các bước thực hiện tìm kiếm nhị phân:

- Xác định giá trị trung bình của danh sách. 
- Xác định xem giá trị cần tìm có lớn hơn hay nhỏ hơn giá trị trung bình đó.
- Nếu giá trị cần tìm lớn hơn giá trị trung bình, thực hiện lại các bước trên với nửa phía sau của danh sách.
- Nếu giá trị cần tìm nhỏ hơn giá trị trung bình, thực hiện lại các bước trên với nửa phía trước của danh sách.
- Lặp lại các bước trên cho đến khi tìm thấy phần tử cần tìm hoặc xác định không có phần tử đó trong danh sách.
- Tìm kiếm nhị phân là một thuật toán hiệu quả vì thao tác cắt giảm danh sách đi một nửa tại mỗi lần lặp lại, vì vậy số lần lặp - lại cần thiết để tìm kiếm phần tử cần tìm trong danh sách lớn sẽ ít hơn so với các thuật toán tìm kiếm khác.


```python
def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    mid = 0

    while low <= high:

        mid = (high + low) // 2

        # Kiểm tra xem x có ở giữa hay không
        if arr[mid] < x:
            low = mid + 1

         # Nếu x nhỏ hơn giá trị ở giữa thì tìm trong nửa trái của danh sách
        elif arr[mid] > x:
            high = mid - 1

         # Nếu x bằng giá trị ở giữa thì trả về chỉ số của phần tử đó
        else:
            return mid

    # Nếu không tìm thấy x trong danh sách, trả về -1
    return -1
```

```python
arr = [2, 3, 4, 10, 40] # danh sách đã được sắp xếp trước
x = 10

result = binary_search(arr, x)

if result != -1:
    print("Phần tử đã được tìm thấy tại chỉ số", str(result))
else:
    print("Không tìm thấy phần tử trong danh sách")
```
