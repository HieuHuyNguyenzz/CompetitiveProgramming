# Binary Search
Tm kim nh phn l mt thut ton c s dng  tm kim phn t trong mt danh sch  c sp xp. Thut ton ny hot ng bng cch lp i lp li qu trnh ct gim na phn khng cn thit ca danh sch cho n khi tm ra phn t cn tm hoc xc nh rng phn t  khng c trong danh sch.

Cc bc thc hin tm kim nh phn:

- Xc nh gi tr trung bnh ca danh sch. 
- Xc nh xem gi tr cn tm c ln hn hay nh hn gi tr trung bnh .
- Nu gi tr cn tm ln hn gi tr trung bnh, thc hin li cc bc trn vi na pha sau ca danh sch.
- Nu gi tr cn tm nh hn gi tr trung bnh, thc hin li cc bc trn vi na pha trc ca danh sch.
- Lp li cc bc trn cho n khi tm thy phn t cn tm hoc xc nh khng c phn t  trong danh sch.
- Tm kim nh phn l mt thut ton hiu qu v thao tc ct gim danh sch i mt na ti mi ln lp li, v vy s ln lp - li cn thit  tm kim phn t cn tm trong danh sch ln s t hn so vi cc thut ton tm kim khc.


```python
def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    mid = 0

    while low <= high:

        mid = (high + low) // 2

        # Kim tra xem x c  gia hay khng
        if arr[mid] < x:
            low = mid + 1

         # Nu x nh hn gi tr  gia th tm trong na tri ca danh sch
        elif arr[mid] > x:
            high = mid - 1

         # Nu x bng gi tr  gia th tr v ch s ca phn t 
        else:
            return mid

    # Nu khng tm thy x trong danh sch, tr v -1
    return -1
```

```python
arr = [2, 3, 4, 10, 40] # danh sch  c sp xp trc
x = 10

result = binary_search(arr, x)

if result != -1:
    print("Phn t  c tm thy ti ch s", str(result))
else:
    print("Khng tm thy phn t trong danh sch")
```
